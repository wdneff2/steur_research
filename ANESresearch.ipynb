```python
import pandas as pd
df = pd.read_csv("anes_pilot_2018.csv")
df['turnout18'] = df["turnout18"].replace([2,3,4],[1,1,2])
df = df[(df['turnout18'] == 1)| (df['turnout18'] == 2)]
df['race'] = df["race"].replace([1,2,3,4,5,6,7],['White','Black','Hispanic', "Asain", "Native American", "Mixed", 'Other'])
df = df.reset_index()
df['race'].unique()
```

```python
import statsmodels.formula.api as smf
results = smf.ols('follow ~ gesad', data=df).fit()
results.summary()
```

```python
results = smf.ols('follow ~ gesad + pid7 + gender + race+ educ', data=df).fit()
results.summary()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X = df[['gesad']]
y = df['turnout18']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
log_regression = LogisticRegression()
log_regression.fit(X_train,y_train)
y_pred = log_regression.predict(X_test)
print(log_regression.intercept_, log_regression.coef_,)
```

```python
def correctanswer(x):
    numbercorrect = 0
    sentence1 = str(df['pk_cjus'][x])
    word1 = 'justice'
    if word1 in sentence1:
        numbercorrect = numbercorrect + 1
    sentence2 = str(df['pk_germ'][x])
    word2 = 'german'
    if word2 in sentence2:
        numbercorrect = numbercorrect + 1
    if 1 == df['pk_spend'][x]:
        numbercorrect = numbercorrect + 1
    if 1896 == df['pk_geer'][x]:
        numbercorrect = numbercorrect + 1
    if 1867 == df['pk_alaska'][x]:
        numbercorrect = numbercorrect + 1
    return numbercorrect
```

```python
data = []
for i in range(len(df)):
    numbercorrect = correctanswer(i)
    d = numbercorrect
    data.append(d)
df["numbercorrect"] = data
# knowledge = pd.DataFrame(data)
# knowledge
df
```

```python
results = smf.ols('numbercorrect ~ gesad', data=df).fit()
results.summary()
```
